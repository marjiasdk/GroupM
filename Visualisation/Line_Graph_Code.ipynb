```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
iterations = [777, 2520, 10948, 23075, 35158, 46574, 101903]
tree = [1.000, 0.993, 0.995, 0.989, 0.990, 0.986, 0.963]
svm = [1.000, 0.984, 0.986, 0.969, 0.964, 0.956, 0.910]
knn = [1.000, 0.985, 0.986, 0.973, 0.967, 0.960, 0.919]

index = np.arange(len(iterations))

bar_width = 0.15

plt.figure(figsize=(10, 6))

plt.bar(index - bar_width, tree, bar_width, label='Decision Trees', color='slategrey')
plt.bar(index, svm, bar_width, label='SVM', color='turquoise')
plt.bar(index + bar_width, knn, bar_width, label='K-NN', color='navy')
plt.xlabel('Dataset Size')
plt.ylabel('Training Accuracy')

plt.xticks(index, iterations)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.ylim(0,1.0)

plt.legend(loc='center right')

plt.tight_layout()
plt.savefig("barchart1.png",bbox_inches='tight')

plt.show()
```
