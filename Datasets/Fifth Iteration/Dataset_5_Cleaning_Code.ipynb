```python
import numpy as np
import pandas as pd
from textblob import Word
```

```python
df_1 = pd.read_csv("symptomsDisease246k.csv")

df_1
```

```python
word_counts = df_1['response'].value_counts()

minimum = 1100
maximum = 1220
filtered_words = word_counts[(word_counts > minimum) & (word_counts < maximum)]

(filtered_words)
```

```python
df_1['response'] = df_1['response'].str.strip()
```

```python
classes = ['response', 'cystitis', 'vulvodynia', 'nose disorder', 'complex regional pain syndrome', 'spondylosis', 'hypoglycemia', 
           'peripheral nerve disorder', 'esophagitis', 'vaginal cyst', 'conjunctivitis due to allergy', 'diverticulitis',
           'gastrointestinal hemorrhage', 'acute bronchitis', 'pneumonia', 'fungal infection of the hair', 'infectious gastroenteritis',
           'spontaneous abortion', 'sprain or strain', 'gout', 'arthritis of the hip', 'strep throat', 'marijuana abuse', 
           'injury to the arm', 'liver disease', 'bursitis', 'eczema', 'spinal stenosis', 'acute pancreatitis', 'injury to the leg']
df_2 = df_1[df_1['response'].isin(classes)]

df_2
```

```python
df_3 = df_2.reset_index(drop=True)
df_3.dropna()
df_3
```

```python
misspelled_words = df_3[df_3['response'] != df_3['response'].apply(lambda x: Word(x).correct())]

# Count the number of misspelled words
misspelled_count = misspelled_words['response'].nunique()

print("Number of unique misspelled words:", misspelled_count)
```

```python
df_3.columns.values[0] = 'Symptom'
df_3.columns.values[1] = 'Disease'
```
