```python
df_1['response'] = df_1['response'].str.strip()
```

```python
classes = ['parkinson disease', 'dysthymic disorder', 'eating disorder', 'social phobia', 'adjustment reaction']
df_2 = df_1[df_1['response'].isin(classes)]

df_2
```

```python
df_3 = df_2.reset_index(drop=True)
df_3.dropna()
df_3
```

```python
misspelled_words = df_3[df_3['response'] != df_3['response'].apply(lambda x: Word(x).correct())]

# Count the number of misspelled words
misspelled_count = misspelled_words['response'].nunique()

print("Number of unique misspelled words:", misspelled_count)
```

```python
df_3.columns.values[0] = 'Symptom'
df_3.columns.values[1] = 'Disease'
```
